# CAMPUS RECRUITMENT

#### DECLARAÇÃO DO PROBLEMA

- Este conjunto de dados consiste em dados de colocação de alunos em um campus XYZ. Inclui percentual e especialização do ensino médio e superior. Também inclui especialização de graduação, tipo e experiência de trabalho e ofertas salariais para os alunos colocados

#### DICIONÁRIO DE DADOS

- sl_no : Serial Number
- gender: Gender- Male='M',Female='F'
- ssc_p : Secondary Education percentage- 10th Grade
- ssc_b : Board of Education- Central/ Others
- hsc_b : Higher Secondary Education percentage- 12th Grade
- hsc_s : Specialization in Higher Secondary Education
- degree_p: Degree Percentage
- degree_t: Under Graduation(Degree type)- Field of degree education
- workex : Work Experience
- etest_p: Entrance Test Percentage
- mba_p: MBA Percentage
- status : Placed or not
- salary : Salary offered

```python
# Outliers

def outliers(data, label):
    plt.figure(figsize = (5,5))
    plt.style.use('seaborn-white')
    ax = plt.subplot()
    plt.boxplot(data)
    ax.set_title(label)
```

```python
outliers(df['ssc_p'], 'Porcentagem do ensino médio')
```

```python
outliers(df['hsc_p'], 'Porcentagem de ensino médio superior')
```

```python
outliers(df['degree_p'], 'Porcentagem de Grau UG')
```

```python
outliers(df['etest_p'], 'Percentual de empregabilidade')
```

- Como podemos ver, temos um número muito menor de outliers em nossos recursos. Especialmente, temos a maioria dos valores discrepantes na porcentagem de hsc_p. Vamos esclarecê-los!

```python
Q1 = df['hsc_p'].quantile(0.25)
Q3 = df['hsc_p'].quantile(0.75)
IQR = Q3 - Q1

filter = (df['hsc_p'] >= Q1 - 1.5 * IQR) & (df['hsc_p'] <= Q3 + 1.5 * IQR)
df_filtered = df.loc[filter]

plt.figure(figsize = (15,5))
plt.style.use('seaborn-white')
ax = plt.subplot(121)
plt.boxplot(df['hsc_p'])
ax.set_title('Antes de remover outliers(hsc_p)')
ax = plt.subplot(122)
plt.boxplot(df_filtered['hsc_p'])
ax.set_title('Depois de remover outliers (hsc_p)')
```
